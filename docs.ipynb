```python
import tempfile
import os
from langchain.document_loaders.unstructured import UnstructuredFileLoader
from typing import List

def load_and_split_documents(file_paths: List[str]):
    documents = []
    for file_path in file_paths:
        loader = UnstructuredFileLoader(file_path, mode="elements")  # You can add other kwargs here
        docs = loader.load()
        documents.extend(docs)
    return documents
```

```python
# Load the file and then convert it to bytes
file_path = "./garden_guru.xlsx"
with open(file_path, "rb") as f:
    bytes = f.read()
# Create a temporary file
temp_file = tempfile.NamedTemporaryFile(delete=False)
temp_file.write(bytes)
loader = UnstructuredFileLoader(temp_file, mode="elements")
docs = loader.load()

```
